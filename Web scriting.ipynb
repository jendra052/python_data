```python
# importing requests and beautiful soup which are already installed in my local computer 

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# requests library send http get to url and retrieve 
# Beautiful class from bs4 parse the HTML content 

url = "https://en.wikipedia.org/wiki/List_of_bank_failures_in_the_United_States_(2008%E2%80%93present)"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
print(soup)
```

```python
# this code find the table element , tr tag is to find row of the table (tr)
# find all is to find all <td>
# since col 1 is for bank name , it will print bank name 

table = soup.find("table", class_="wikitable sortable")
rows = table.find_all("tr")

for row in rows:
    cols = row.find_all("td")
    if cols:
        a_elem = cols[1].find("a")
        if a_elem:
            bank_name = a_elem.text
            print(bank_name)
```

```python
#looking for' a' elements in above code gives bank name with link , there are only 4 banks in 2008 table 
# so we can try to extract bank name with <td> element 


table = soup.find("table", class_="wikitable sortable")
rows = table.find_all("tr")

for row in rows:
    cols = row.find_all("td")
    if cols and cols[1]:
        bank_name = cols[1].text.strip()
        print(bank_name)
```

```python
url = "https://en.wikipedia.org/wiki/List_of_bank_failures_in_the_United_States_(2008%E2%80%93present)"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

data = []
for table in soup.find_all("table", class_="wikitable sortable"):
    for row in table.find_all("tr"):
        cols = row.find_all("td")
        if len(cols) >= 5:
            bank_name = cols[1].text.strip()
            city = cols[2].text.strip()
            state = cols[3].text.strip()
            closing_date = cols[4].text.strip()
            acquiring_institution = cols[5].text.strip()
            estimated_loss = ""
            if len(cols) >= 6:
                estimated_loss = cols[6].text.strip()
            data.append([bank_name, city, state, closing_date, acquiring_institution, estimated_loss])

df = pd.DataFrame(data, columns=["Bank Name", "City", "State", "Closing Date", "Acquiring Institution", "Estimated Loss"])
print(df)
```

#  fuzzy matching - bank name vs acquiring institution 

```python
from fuzzywuzzy import fuzz

# Define a function to compute the similarity score between two strings
def compute_similarity(row):
    name1 = row["Bank Name"]
    name2 = row["Acquiring Institution"]
    return fuzz.token_set_ratio(name1, name2)

# Apply the function to each row in the DataFrame to compute the similarity scores
df["Similarity Score"] = df.apply(compute_similarity, axis=1)

# Sort the DataFrame by the similarity scores in descending order
df = df.sort_values(by="Similarity Score", ascending=False)

# Print the top 10 rows of the DataFrame with the highest similarity scores
print(df.head(10))
```
