```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('banklist.csv')
print(df)
df.head(20)
```

```python
# Create a dictionary to map the old header names to the new header names
header_map = {
    'Bank Name': 'Name',
    'City': 'City Name',
    'State': 'STATE',
    'Cert': 'Cert Number',
    'Acquiring Institution': 'Acquirer',
    'Closing Date': 'Close Date',
    'Fund': 'Funding'
}

# Rename the headers using the dictionary
df = df.rename(columns=header_map)

# Print the first few rows of the DataFrame to verify the header names have changed
print(df.head())
```

```python
# checking structures of data set by using info()and describes() methods

df.info()
df.describe()
```

```python
#Check for missing values using the isnull() and sum() methods.

df.isnull().sum()
```

```python
# there is no missing value in the data set, incase missing value it we can use
# drop column methods to drop column with missing value for example column with 50% missing value 

# Drop columns with more than 50% missing values
threshold = len(df) * 0.5
df.dropna(thresh=threshold, axis=1, inplace=True)

# Fill missing values in remaining columns with the mean or mode

df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# I check if there is any changes after missing value and replacing missing vlue mthods 

df.head(10)
```

```python
#Check for duplicates using the duplicated() and sum() methods.
df.duplicated().sum()
```

```python
# there is no dublicate , if in case dublicate we can use drop_deblicate methods 

df.drop_duplicates(inplace=True)
```

```python
# df.dtype method will provides information data type in each column

df.dtypes
```

```python
# two match into 2 digit year format 

df['Closing Date'] = pd.to_datetime(df['Closing Date'], format="%d-%b-%y")
```

```python
# fuzzywuzzy: This library provides several methods for fuzzy matching,
#including the fuzzy_match method for finding the best match between two strings.

# i am using Bank Name column string to match

import pandas as pd
df = pd.read_csv('cleaned_banklist.csv')
from fuzzywuzzy import fuzz, process

# create a list of unique bank names
bank_names = df['Bank Name'].unique()

# create a function to find the best match for a given bank name
def find_best_match(name, choices):
    return process.extractOne(name, choices, scorer=fuzz.token_sort_ratio)

# test the function with a sample name
name = 'The First State Bank'
match, score, idx = find_best_match(name, bank_names)
print(f"Match: {match}, Score: {score}, Index: {idx}")
```

```python
#find_best_match function is not being called 


from fuzzywuzzy import fuzz, process

# create a list of unique bank names
bank_names = df['Bank Name'].unique()

# create a function to find the best match for a given bank name
def find_best_match(name, choices):
    best_match = None
    best_score = 0
    best_idx = None
    for idx, choice in enumerate(choices):
        score = fuzz.ratio(name.lower(), choice.lower())
        if score > best_score:
            best_match = choice
            best_score = score
            best_idx = idx
    return best_match, best_score, best_idx

# test the function with a sample name
name = 'City National Bank'
match, score, idx = find_best_match(name, bank_names)
print(f"Match: {match}, Score: {score}, Index: {idx}")
```
